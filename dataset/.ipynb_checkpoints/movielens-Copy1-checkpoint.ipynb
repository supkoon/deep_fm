## 2.Wide part

* embedding V = 15
* total_feature = 108
* total_grouped_feature = 9

```python
class wide_part(keras.layers.Layer):
    def __init__(self,V,**kwargs):
        super().__init__(**kwargs)
        #order - 1 weight
        self.W = keras.layers.Dense(1)
        self.V = V
        self.Embedding_output = []
        
    def call(self,continuous_input,categorical_input):
        #order - 1 
        concat_order_1 = keras.layers.concatenate(continuous_input+categorical_input)
        order_1_output = self.W(concat_order_1)
        

        #order - 2
        Embedding_layers = [keras.layers.Embedding(field.shape[1], self.V) for field in continuous_input+categorical_input]
        for index,field in enumerate(continuous_input+categorical_input):
#             print(field)
            result = Embedding_layers[index](field)
#             print(result)
            #(None,N) --> #(None,15)
            result = keras.layers.Lambda(lambda x: tf.reduce_mean(x, axis = 1, keepdims=True))(result)
#             print(result)
            self.Embedding_output.append(result)
        
        
        #concat_embedding_output
        #(None,135)
        embed_2d = keras.layers.concatenate(self.Embedding_output,axis=1)
        
        
        tensor_sum = keras.layers.Lambda(lambda x: keras.backend.sum(x, axis = 1), name = 'sum_of_tensors')
        tensor_square = keras.layers.Lambda(lambda x: keras.backend.square(x), name = 'square_of_tensors')

        sum_of_embed = tensor_sum(embed_2d)
        square_of_embed = tensor_square(embed_2d)

        square_of_sum = keras.layers.Multiply()([sum_of_embed, sum_of_embed])
        sum_of_square = tensor_sum(square_of_embed)

        sub = keras.layers.Subtract()([square_of_sum, sum_of_square])
        sub = keras.layers.Lambda(lambda x: x*0.5)(sub)
        order_2_output = keras.layers.Reshape((1,), name = 'fm_2d_output')(tensor_sum(sub))

        wide_output = keras.layers.concatenate([order_1_output,order_2_output])
        
        return wide_output, embed_2d

        
```

## 3.Deep part

in paper,

(1) dropout: 0.5;   
(2) network structure: 400-400-400;  
(3) opti- mizer: Adam;   
(4) activation function: relu 

```python
class deep_part(keras.layers.Layer):
    def __init__(self,layer_list = [400,400,400], dropout_rate = 0.5, activation = "relu", **kwargs):
        super().__init__(**kwargs)
        self.activaiton_fn = keras.activations.get(activation)
        self.dropout_rate = dropout_rate
        self.dense_layer_list = [keras.layers.Dense(num_neuron, activation = self.activaiton_fn) for num_neuron in layer_list]
        self.output_layer  = keras.layers.Dense(1, activation  = "relu")

    def call(self,inputs):
        embed_2d = inputs
        embed_2d = keras.layers.Flatten(name = 'flat_embed')(embed_2d)

        result = embed_2d
        for layer in self.dense_layer_list:
            result = keras.layers.Dropout(self.dropout_rate)(result)
            result = layer(result)
            
        deep_result = self.output_layer(result)
        
        return deep_result
         
```

```python
class deep_fm_model(keras.Model):
    def __init__(self,V=15,layer_list = [400,400,400], dropout_rate = 0.5, activation = "relu",**kwargs):
        super().__init__(**kwargs)
        self.V = V
        self.wide_part = wide_part(self.V)
        self.deep_part = deep_part(layer_list,dropout_rate,activation)
        self.output_layer = keras.layers.Dense(1,activation = "sigmoid")
        
        
    def call(self,inputs):
        continuous_input, categorical_input = inputs 
        #make_input_layers
        continuous_Input_layers,categorical_Input_layers = make_input_layers(continuous_input, categorical_input)
        #wide_part
        wide_output, embeddings =self.wide_part(continuous_Input_layers,categorical_Input_layers)

        #deep_part
        deep_output = self.deep_part(embeddings)
        
        #concat_wide_deep
        wide_deep_output = keras.layers.concatenate([wide_output,deep_output])
        wide_deep_output = self.output_layer(wide_deep_output)

        return wide_deep_output
        
        
```
